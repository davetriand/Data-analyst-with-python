```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
df = pd.read_csv('mallCustomerData.txt', sep=",")
print(df.shape)
df.head(10)
```

```python
print(df['Gender'].value_counts())
```

```python
#getting the values and plotting it
f1 = df['Annual Income (k$)'].values
f2 = df['Spending Score (1-100)'].values

for key in df.keys():
    print(key)

X = np.array(list(zip(f1, f2)))
#X=[[1,1], [1,2], [3,3], [4,4]]
plt.scatter(f1, f2, c='black', s=20)
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
C = kmeans.cluster_centers_
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 1])
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 1], c = 'y')
ax.scatter(C[:, 0], C[:, 1], C[:, 1], marker='*', c='#050505', s=1000)
```

```python
# nitializing Kmeans
kmeans = KMeans(n_clusters=4)
# Fitting the input data
kmeans = kmeans.fit(X)
# predict the cluster for each data point
labels = kmeans.predict(X)
# getting the cluster centers
C = kmeans.cluster_centers_
print(labels)

```
