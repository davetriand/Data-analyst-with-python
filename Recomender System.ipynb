```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['number of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10, 4))
ratings['number of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10, 4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='number of ratings', data=ratings, alpha=0.5)
```

```python
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
moviemat.head()
```

```python
ratings.sort_values(by='number of ratings', ascending=False).head(10)
```

```python
StarWars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
StarWars_user_ratings.head()
```

```python
similar_to_StarWars = moviemat.corrwith(StarWars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_StarWars = pd.DataFrame(similar_to_StarWars, columns=['Correlation'])
corr_StarWars.dropna(inplace=True)
corr_StarWars.head()
```

```python
corr_StarWars.sort_values(by='Correlation', ascending=False).head(10)
```

```python
corr_StarWars = corr_StarWars.join(ratings['number of ratings'])
corr_StarWars.head()
```

```python
corr_StarWars[corr_StarWars['number of ratings'] > 100].sort_values(by='Correlation', ascending=False).head(10)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['number of ratings'])
corr_liarliar[corr_liarliar['number of ratings'] > 100].sort_values(by='Correlation', ascending=False).head(10)
```
